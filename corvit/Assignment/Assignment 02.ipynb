```python
users = {"admin": "1234"}

username = st.text_input("Username")
password = st.text_input("Password", type="password")

if st.button("Login"):
    if username in users and users[username] == password:
        st.success("Login Successful")
    else:
        st.error("Invalid Credentials")
```

2. Student Management (OOP)

Use class + objects

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.__gpa = 0   # encapsulation

    def set_gpa(self, gpa):
        self.__gpa = gpa

    def get_gpa(self):
        return self.__gpa
```

3. Course Management (Inheritance)

```python
class Course:
    def __init__(self, title):
        self.title = title

class ProgrammingCourse(Course):
    def __init__(self, title, language):
        super().__init__(title)
        self.language = language
```

4. Grade Calculation Logic

```python
def calculate_gpa(marks):
    if marks >= 85:
        return 4.0
    elif marks >= 70:
        return 3.0
    elif marks >= 50:
        return 2.0
    else:
        return 0.0
```
