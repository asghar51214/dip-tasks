Searching Methods
1.find()

```python
text = "Hi python"
print(text.find("python"))   # 6
```

 rfind()

```python
text = "apple mango apple"
print(text.rfind("apple"))   # Last position
```

index()

```python
text = "Hello World"
print(text.index("World"))   # 6
```

rindex()

```python
text = "apple mango apple"
print(text.rindex("apple"))  # Last position
```

Replacing Methods replace()

```python
text = "I like Java"
print(text.replace("Java", "Python"))
```

translate()

```python
text = "abc"
table = str.maketrans("a", "x")
print(text.translate(table))
```

Splitting and Joining
split()

```python
text = "apple,banana,mango"
print(text.split(","))
```

rsplit()

```python
text = "apple,banana,mango"
print(text.rsplit(",", 1))
```

partition()

```python
text = "apple-banana"
print(text.partition("-"))
```

rpartition()

```python
text = "apple-banana-orange"
print(text.rpartition("-"))
```

splitlines()

```python
text = "Hello\nWorld"
print(text.splitlines())
```

join()

```python
words = ["Hello", "World"]
print(" ".join(words))
```
